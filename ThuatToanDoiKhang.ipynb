I.	Thuật toán đối kháng
Thuật toán tìm kiếm đối khánglà một quyết định tối ưu trong trò chơi. Nghĩa là tìm một đường đi bảo đảm chiến thắng cho người chơi.
-	Các thành phần
•	Trạng thái ban đầu (initial state): Trạng thái của trò chơi + Người chơi nào được đi nước đầu tiên
•	Trạng thái kết thúc (terminal : Kiểm tra kết thúc trò chơi
•	chuyển trạng thái (Sucessor)Trả về thông tin (nước đi, trạng thái). Nghĩa là nước đi hợp lệ từ trạng thái hiện tại, trạng thái mới được chọn
•	Hàm lợi ích : đánh giá trạng thái kế
•	Hàm lượng giá heuristic E(n) = X(n) –O(n) với n số khả năng thắng của quân X và số khả năng thắng của quân O


II.	Thuật toán Minimax
-	Minimax (còn gọi là minmax) là một phương pháp trong lý thuyết quyết định có mục đích là tối thiểu hóa (minimize) tổn thất vốn được dự tính có thể là "tối đa" (maximize). Có thể hiểu ngược lại là, nó nhằm tối đa hóa lợi ích vốn được dự tính là tối thiểu (maximin). Nó bắt nguồn từ trò chơi có tổng bằng không. Nó cũng được mở rộng cho nhiều trò chơi phức tạp hơn và giúp đưa ra các quyết định chung khi có sự hiện diện của sự không chắc chắn.
//"Tổng bằng không" có nghĩa là lợi ích của người chơi này là mất mát của người chơi kia.
-	Trong những trò chơi này, mỗi người chơi đều muốn thực hiện nước đi tốt nhất hoặc tối ưu nhất để tối đa hóa cơ hội chiến thắng của mình, đồng thời cũng cố gắng giảm thiểu cơ hội thắng của đối thủ. Thuật toán này giúp máy tính hoặc AI quyết định nước đi tiếp theo bằng cách xem xét tất cả các nước đi khả thi mà nó có thể thực hiện và mọi phản ứng có thể từ đối thủ. Nó thực hiện điều này bằng cách mô phỏng các nước đi và kết quả tương lai của trò chơi.
-	Triết lý hoạt động

•	MAX (Người chơi tối đa hóa): Đây là người chơi mà chúng ta (AI) đang điều khiển. Mục tiêu của MAX là chọn một nước đi để tối đa hóa điểm số của mình, giả định rằng đối thủ sẽ chơi để tối thiểu hóa điểm số đó.
•	MIN (Người chơi tối thiểu hóa): Đây là đối thủ. Mục tiêu của MIN là chọn một nước đi để tối thiểu hóa điểm số của MAX.

III.	Chiến lược cắt tỉa Beta/Alpha
Thuật toán Alpha-Beta Pruning là một phương pháp tối ưu hóa để giảm số lượng nút kiểm tra không cần thiết trong thuật toán Minimax. Thuật toán này đặt các giới hạn (alpha và beta) cho giá trị tốt nhất của các nút. Nếu nút hiện tại nằm ngoài khoảng giữa alpha và beta, các nút con của nó sẽ không được kiểm tra. Điều này giúp tăng tốc độ tìm kiếm và giảm độ phức tạp tính toán.
Cắt tỉa Alpha-Beta được phát minh như một giải pháp tối ưu hóa cho Minimax. Nó xuất phát từ sự nhận biết tinh tế rằng: "Nếu một nước đi đã tỏ ra tồi tệ hơn một lựa chọn khác mà ta đã có, thì không cần phải tốn thời gian phân tích nước đi đó sâu hơn nữa."

Các thành phần chính: Alpha và Beta
•	Alpha (α): Đại diện cho giá trị tốt nhất (cao nhất) mà người chơi MAX  có thể đảm bảo nhận được trên đường đi từ gốc đến nút hiện tại. Ban đầu, Alpha được khởi tạo với giá trị âm vô cùng (−∞).
•	Beta (β): Đại diện cho giá trị tốt nhất (thấp nhất) mà người chơi có thể đảm bảo nhận được trên đường đi từ gốc đến nút hiện tại. Ban đầu, Beta được khởi tạo với giá trị dương vô cùng (+∞).
Khi thuật toán đi qua cây, nó cập nhật α và β. Nếu phát hiện một nhánh có giá trị tệ hơn α hoặc β, thì cắt bỏ nhánh đó, vì nó không thể ảnh hưởng đến quyết định cuối cùng.

1. Mô tả Code
Phần đầu tiên của mã nguồn thực hiện việc nhập hai thành phần chính từ thư viện Tkinter:
import tkinter as tk và from tkinter import messagebox.
Điều này cho phép tạo cửa sổ ứng dụng và hiển thị các hộp thoại thông báo kết quả một cách trực quan.
Lớp TicTacToe là trung tâm của chương trình. Trong phương thức khởi tạo init(self), các thành phần giao diện và trạng thái ban đầu được thiết lập một cách có hệ thống:

Một cửa sổ chính được tạo với tiêu đề “Trò chơi Tic Tac Toe - X và O”, đồng thời bị khóa kích thước để đảm bảo bố cục luôn ổn định.
Biến self.current_player được khởi tạo với giá trị True để biểu thị người chơi X đi trước.
Ma trận logic self.board được tạo dưới dạng danh sách lồng nhau 3×3, ban đầu tất cả các ô đều mang giá trị None.
Chín nút Button được tạo và sắp xếp theo lưới 3×3 bằng phương thức grid. Mỗi nút có kích thước và phông chữ lớn để dễ quan sát, đồng thời được gán sự kiện nhấn thông qua lệnh command=lambda row=i, col=j: self.button_click(row, col). Việc sử dụng lambda với tham số mặc định là kỹ thuật quan trọng nhằm tránh lỗi late binding thường gặp trong Python, đảm bảo tọa độ ô được truyền chính xác khi người dùng nhấn.
Một nhãn Label hiển thị lượt chơi hiện tại và một nút “Chơi lại” được đặt bên dưới bàn cờ để hoàn thiện giao diện.

Phương thức button_click(self, row, col) là nơi xử lý toàn bộ logic của một lượt đi. Khi người chơi nhấn vào một ô, hàm này thực hiện tuần tự các bước sau:

Kiểm tra tính hợp lệ của nước đi: nếu ô đã được đánh hoặc trò chơi đã có người thắng thì bỏ qua.
Xác định ký hiệu hiện tại (“X” hoặc “O”) dựa trên biến self.current_player, đồng thời áp dụng màu chữ tương ứng (đỏ cho X, xanh lá cho O).
Cập nhật ký hiệu lên nút và ghi nhận trạng thái vào ma trận self.board.
Gọi hàm kiểm tra chiến thắng. Nếu có người thắng, hiển thị thông báo bằng messagebox và tô vàng đường thắng.
Nếu không có người thắng nhưng bàn cờ đã đầy, thông báo kết quả hòa.
Nếu trò chơi vẫn tiếp diễn, chuyển lượt chơi và cập nhật nội dung cũng như màu sắc của nhãn hướng dẫn.

Hàm check_winner(self) chịu trách nhiệm kiểm tra tám trường hợp chiến thắng có thể xảy ra: ba hàng ngang, ba cột dọc và hai đường chéo. Điểm nổi bật của hàm này là không chỉ trả về giá trị boolean mà còn trả về cặp tọa độ đầu-cuối của đường thắng (nếu tồn tại). Thiết kế này cho phép phương thức highlight_winner(self) sau đó xác định chính xác các ô cần tô màu nền vàng, mang lại hiệu ứng hình ảnh sinh động và dễ nhận biết.
Các phương thức phụ trợ khác bao gồm:

is_full(self): kiểm tra xem bàn cờ đã kín chưa để xác định kết quả hòa.
reset_game(self): đặt lại toàn bộ trạng thái về ban đầu – đưa lượt về X, xóa ký hiệu và màu nền của tất cả các nút, đồng thời cập nhật lại nhãn lượt chơi.
run(self): khởi động vòng lặp sự kiện chính của Tkinter thông qua lệnh mainloop().

Cuối cùng, khối lệnh if name == "main": đảm bảo khi file được chạy trực tiếp, một đối tượng TicTacToe sẽ được tạo và trò chơi bắt đầu ngay lập tức.

import tkinter as tk
from tkinter import messagebox

class TicTacToe:
    def __init__(self):
        self.window = tk.Tk()
        self.window.title("Trò chơi Tic Tac Toe - X và O")
        self.window.resizable(False, False)
        
        # Người chơi hiện tại: True = X, False = O
        self.current_player = True  # Bắt đầu bằng X
        
        # Ma trận 3x3 lưu trạng thái ô (None, "X", "O")
        self.board = [[None]*3 for _ in range(3)]
        
        # Tạo 9 nút
        self.buttons = [[None]*3 for _ in range(3)]
        for i in range(3):
            for j in range(3):
                self.buttons[i][j] = tk.Button(self.window,
                                               text="",
                                               font=("Arial", 30, "bold"),
                                               width=5,
                                               height=3,
                                               bg="#f0f0f0",
                                               command=lambda row=i, col=j: self.button_click(row, col))
                self.buttons[i][j].grid(row=i, column=j)
        
        # Nhãn hiển thị lượt chơi
        self.label = tk.Label(self.window,
                              text="Lượt của: X",
                              font=("Arial", 18),
                              fg="blue")
        self.label.grid(row=3, column=0, columnspan=3, pady=10)
        
        # Nút chơi lại
        self.restart_btn = tk.Button(self.window,
                                     text="Chơi lại",
                                     font=("Arial", 14),
                                     bg="#4CAF50",
                                     fg="white",
                                     command=self.reset_game)
        self.restart_btn.grid(row=4, column=0, columnspan=3, pady=10)
        
    def button_click(self, row, col):
        # Nếu ô đã được đánh hoặc đã có người thắng thì không làm gì
        if self.board[row][col] is not None or self.check_winner():
            return
        
        # Đánh dấu ô
        symbol = "X" if self.current_player else "O"
        color = "red" if self.current_player else "green"
        
        self.board[row][col] = symbol
        self.buttons[row][col].config(text=symbol, fg=color)
        
        # Kiểm tra thắng
        if self.check_winner():
            messagebox.showinfo("Kết quả", f"Người chơi {symbol} thắng!")
            self.highlight_winner()
            return
        
        # Kiểm tra hòa
        if self.is_full():
            messagebox.showinfo("Kết quả", "Hòa!")
            return
        
        # Chuyển lượt
        self.current_player = not self.current_player
        next_player = "X" if self.current_player else "O"
        next_color = "blue" if self.current_player else "darkgreen"
        self.label.config(text=f"Lượt của: {next_player}", fg=next_color)
    
    def check_winner(self):
        # Kiểm tra hàng
        for i in range(3):
            if (self.board[i][0] == self.board[i][1] == self.board[i][2]
                and self.board[i][0] is not None):
                return (i, 0), (i, 2)  # trả về tọa độ đầu và cuối của đường thắng
        
        # Kiểm tra cột
        for j in range(3):
            if (self.board[0][j] == self.board[1][j] == self.board[2][j]
                and self.board[0][j] is not None):
                return (0, j), (2, j)
        
        # Kiểm tra đường chéo chính
        if (self.board[0][0] == self.board[1][1] == self.board[2][2]
            and self.board[0][0] is not None):
            return (0, 0), (2, 2)
        
        # Kiểm tra đường chéo phụ
        if (self.board[0][2] == self.board[1][1] == self.board[2][0]
            and self.board[0][2] is not None):
            return (0, 2), (2, 0)
        
        return None
    
    def highlight_winner(self):
        winner_line = self.check_winner()
        if winner_line:
            (r1, c1), (r2, c2) = winner_line
            # Nếu cùng hàng
            if r1 == r2:
                for c in range(3):
                    self.buttons[r1][c].config(bg="yellow")
            # Nếu cùng cột
            elif c1 == c2:
                for r in range(3):
                    self.buttons[r][c1].config(bg="yellow")
            # Đường chéo
            else:
                for i in range(3):
                    self.buttons[i][i].config(bg="yellow") if r1 == 0 and c1 == 0 else None
                    self.buttons[i][2-i].config(bg="yellow") if r1 == 0 and c1 == 2 else None
    
    def is_full(self):
        for row in self.board:
            if None in row:
                return False
        return True
    
    def reset_game(self):
        self.current_player = True
        self.board = [[None]*3 for _ in range(3)]
        for i in range(3):
            for j in range(3):
                self.buttons[i][j].config(text="", bg="#f0f0f0")
        self.label.config(text="Lượt của: X", fg="blue")
    
    def run(self):
        self.window.mainloop()

# Chạy trò chơi
if __name__ == "__main__":
    game = TicTacToe()
    game.run()
