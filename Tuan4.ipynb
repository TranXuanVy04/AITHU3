import sys

INF = float('inf')

def solve_tsp_dynamic_programming(cost_matrix):

    """
    Gi·∫£i b√†i to√°n Ng∆∞·ªùi b√°n h√†ng (TSP) b·∫±ng Thu·∫≠t to√°n Quy ho·∫°ch ƒë·ªông (Held-Karp).

    cost_matrix: Ma tr·∫≠n chi ph√≠ (kho·∫£ng c√°ch) gi·ªØa c√°c th√†nh ph·ªë.
    
    Tr·∫£ v·ªÅ: (min_cost, optimal_path)
    """
    N = len(cost_matrix)
    if N == 0:
        return 0, []
    if N == 1:
        return 0, [0]

    dp = [[INF] * N for _ in range(1 << N)]

    path_store = [[-1] * N for _ in range(1 << N)]

    start_city = 0
    dp[1 << start_city][start_city] = 0

    for mask in range(1, 1 << N):
        for j in range(N):

            if (mask & (1 << j)):

                previous_mask = mask ^ (1 << j)

                if previous_mask == 0:
                    continue

                for i in range(N):
                    if (previous_mask & (1 << i)):
                        new_cost = dp[previous_mask][i] + cost_matrix[i][j]
                        
                        if new_cost < dp[mask][j]:
                            dp[mask][j] = new_cost
                            path_store[mask][j] = i

    final_mask = (1 << N) - 1
    min_cost = INF
    last_city = -1

    for j in range(1, N):
        total_cost = dp[final_mask][j] + cost_matrix[j][start_city]
        
        if total_cost < min_cost:
            min_cost = total_cost
            last_city = j

    path = []
    current_mask = final_mask
    current_city = last_city

    while current_city != -1:
        path.append(current_city)

        previous_city = path_store[current_mask][current_city]
        
        current_mask = current_mask ^ (1 << current_city)
        current_city = previous_city

    path.append(start_city)
    path.reverse()

    if N > 1:
        path.append(start_city)
    
    return min_cost, path
